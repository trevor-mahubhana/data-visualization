```python
fake_data.age.plot(kind = 'hist',title = 'Distribution of Age',color = 'skyblue',bins = [20,30,40,50,60,70])
plt.axvline(fake_data.age.mean(),label = 'Age Mean',color = 'black',linewidth = 3)
plt.axvline(fake_data.age.median(),label = 'Age Mean',color = 'green',linewidth = 3)
plt.xlabel("Age")
plt.legend(loc='lower right')
plt.show()
```

```python
fake_data.salary.plot(kind = 'hist',title = 'Distribution of Salary',bins = [800,1000,1200,1400,1600,1800,2000,2200],color = 'grey')
plt.xlabel("Salary in $")
plt.axvline(fake_data.salary.mean(),label = 'Mean Salary',color = 'red',linewidth = 3)
plt.axvline(fake_data.salary.median(),label = 'Median Salary',color = 'black',linewidth = 3)
plt.legend(loc = 'upper right')
plt.show()
```

```python
fake_data[['age','salary']].corr()
```

```python
fake_data.plot(kind = 'scatter',x = 'age',y = 'salary',color = 'black',title = 'relationship between age and salary')
plt.show()
```

```python
g_count = fake_data.gender.value_counts()
g_count.plot(kind = 'bar',title= "Distribution of gender",color = ['green','gold'])
plt.ylabel("Count")
plt.xlabel("Gender")
plt.show()
```

```python
edu = fake_data.edu_level.value_counts()
edu.plot(kind = 'barh',title='Distribution of education of our customers',color = ['green','yellow','red'])
plt.ylabel('Count')
plt.xlabel("Educational Level")
plt.show()
```

```python
state_va = fake_data.state.value_counts()
state_va.plot(kind = 'bar',title = "Distribution of States",color = 'skyblue')
plt.ylabel("Count")
plt.xlabel("State")
plt.show()
```

```python
state_va.plot(kind = 'barh',title = "Distribution of States",color = 'khaki')
plt.xlabel("Count")
plt.ylabel("State")
plt.show()
```

```python
state_va.nlargest(5).plot(kind = 'bar',title = "Top 5 States where most of the customers live",color = 'red')
plt.ylabel("Count")
plt.xlabel("State")
plt.show()
```

```python
state_va.nsmallest(5).plot(kind = 'bar',title = "Bottom 5 of States with fewer customers",color = 'green')
plt.ylabel("Count")
plt.xlabel("State")
plt.show()
```

```python
fake_data.age.plot(kind = 'box',color = 'green')
plt.title("5-Number Summary of Age")
plt.show()
```

```python
fake_data.salary.plot(kind = 'box',color = 'green')
plt.title("5-Number Summary of Salary")
plt.show()
```
