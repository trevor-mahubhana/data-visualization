## Creating fake data

```python
def fake_data():
    
    age = np.random.randint(18,70,size = 1000)
    gender = np.random.choice(['male','female'],size = 1000)
    salary = np.random.randint(800,2100,size = 1000)
    edu_level = np.random.choice(['certificates only','undergraduate degree','master degree'],size = 1000)
    state = np.random.choice(['florida','alaska','califonia','washington dc'
                              ,'new york','chicago','denver','miami','san francisco','los angeles'],size = 1000)
    df = pd.DataFrame({'age':age,'gender':gender,'salary':salary,'edu_level':edu_level,'state':state})
    return df

df = fake_data()
np.random.seed(1)
df.to_csv(r"C:\Users\MALCOM_TREVOR\Desktop\DATA SCIENCE\MY PROJECT\data visualization\_fake_dataset.csv",index = False)
```

```python
sns.boxplot(x = 'gender',y = 'salary',data = df,hue = 'edu_level',palette = 'spring')
plt.title('5-number summary for salary')
plt.grid(True)
plt.show()
```

## Heatmap

```python
spearman_corr = df[['age','salary']].corr(method = 'spearman')
```

```python
sns.heatmap(spearman_corr,annot = True,cmap = ['#7FFFD4','#0097A7','#F7DC6F','#DC143C','#ACFFAC','#804A1A'])
plt.title('Heatmap showing the Spearman Correlation Coefficient')
plt.show()
```

```python
pearson_corr = df[['age','salary']].corr(method = 'pearson')
```

```python
sns.heatmap(pearson_corr,annot = True,cmap = ['#4169E1','#C7B8EA','#32CD32','#7c5ce7','#F7DC6F','#DC143C','skyblue'])
plt.title('Heatmap showing the Pearson Correlation Coefficient')
plt.show()
```
