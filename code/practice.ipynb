# creating bar chart and histogram using pandas plotting 

```python
fake.age.plot.hist(bins = [20,30,40,50,60,70],title = ("Distribution of age"),edgecolor = 'red',color = 'silver')
plt.axvline(fake.age.mean(),label = "Mean",color = 'black',linewidth = 3)
plt.axvline(fake.age.median(),label = "Median",color = 'red',linewidth = 3)
plt.legend(title = "key",loc = 'lower right')
plt.xlabel("Age")
plt.show()
```

```python
fake.gender.value_counts().plot.bar(title = "Distribution of gender",color = 'skyblue')
plt.ylabel("Frequency")
plt.show()
```

```python
state_count = fake.state.value_counts()
```

```python
state_count.plot.bar(title="Bar chart showing the distribution of states",width = 0.7)
plt.show()
```

### Creating charts using matplotlib

```python
plt.figure(figsize = (8,6))
plt.hist("age",data = fake ,color = 'skyblue',edgecolor = 'red',bins = [20,30,40,50,60,70])
plt.axvline(fake.age.mean(),label = "Mean",color = 'green')
plt.axvline(fake.age.median(),label = "Median",color = 'black')
plt.legend(title = 'key',loc = 'upper right')
plt.title("Distribution of age")
plt.xlabel("Age")
plt.ylabel("Frequecy")
plt.show()
```

```python
bins = [800,1000,1200,1400,1600,1800,2000]
plt.hist('salary',data = fake ,color = 'yellow',edgecolor = 'green',bins = bins)
plt.axvline(fake.salary.mean(),label = 'Mean',color = 'orange')
plt.axvline(fake.salary.median(),label = 'Median',color = 'gold')
plt.title("Distribution of Salaries")
plt.ylabel("Frequency")
plt.xlabel("salary in $")
plt.legend(title = 'key',loc = 'lower left')
plt.show()
```

```python
# creating pie chart using matplotlib for the gender  and educational level and state columns
```

```python
plt.figure(figsize = (8,8))
pie_gender = fake.gender.value_counts()
plt.pie(pie_gender,autopct ="%.1f%%", pctdistance =0.5,startangle = 180,explode = (0,0),colors = ['orange','blue'])
plt.legend(pie_gender.index,title = "key",loc ="lower left")
plt.show()
```
