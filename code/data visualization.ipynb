### Histograms

```python
random_age = [18,19,21,25,25,32,35,41]
```

```python
for_basic['random_age2'] = random_age
```

```python
for_basic.random_age2
```

```python
ages = for_basic['random_age2']
bins = [10,20,30,40,50] # specify the age want to explore the distribution
plt.figure(figsize=(10,6))
plt.hist(ages,bins=bins,edgecolor = 'green') # number of respondents to be viewed in logarithm,log = True
plt.title('Distribution of age')
plt.xlabel('Ages')
plt.ylabel('Total number of developers')

plt.axvline(ages.median(),color = 'red',label = 'Median age',linewidth = 5 )
plt.axvline(ages.mean(),color = 'black',label = 'Mean age',linewidth = 5 )
plt.legend()
plt.tight_layout()
plt.show()
```

# Creating charts and plots using Seaborn 

## Bar plot

##### works when there is a  categorical variable and a continuous values for instance education level and salaries 
##### there categorical variable is found in the x axis and the continous one is at the y axis
##### use hue  method to specify another categorical variable to be shown in comparison with another on the same plot
##### by default the estimator is a mean and can change by passsing a parameter, estimator = estimator_name

```python
for_basic['Education'] = ['certificate','diploma',"under_grad degree",'diploma','certificate',"master's","under_grad degree","master's"]
```

```python
for_basic
```

```python
sns.barplot(x = 'Education',y = 'python_salary',data = for_basic,color = 'skyblue')
plt.title('Educationa and salaries for Python Developers')
plt.grid(True)
plt.show()
### shows the mean amount  in each category and the amount received
```

```python
sns.barplot(x = 'Education',y = 'python_salary',data = for_basic,color = 'skyblue',saturation = 0.7,ci = 34)
plt.title('Educationa and salaries for Python Developers')
plt.grid(True)
plt.show()
### shows the highest earner in each category and the amount received
```

```python
sns.barplot(x = 'Education',y = 'python_salary',data = for_basic,hue = 'Gender',palette = 'coolwarm')
plt.title('Educationa and salaries for Python Developers')
plt.grid(True)
plt.show()
```

```python
from numpy import median
```

```python
sns.barplot(x = 'Education',y = 'python_salary',data = for_basic,hue = 'Gender',palette = 'spring',estimator = median)
plt.title('Educationa and salaries for Python Developers')
plt.grid(True)
plt.show()
```
